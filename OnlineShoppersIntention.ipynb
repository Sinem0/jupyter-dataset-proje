```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler

# Veri setini yÃ¼kleme
file_path = r'C:\Users\sinem\Desktop\online_shoppers_intention.csv'
data = pd.read_csv(file_path)

# Kategorik verileri one-hot encoding ile sayÄ±sal verilere dÃ¶nÃ¼ÅŸtÃ¼rme
data = pd.get_dummies(data, columns=["Month", "VisitorType"], drop_first=True)

# Hedef deÄŸiÅŸken ve baÄŸÄ±msÄ±z deÄŸiÅŸkenleri ayÄ±rma
X = data.drop("Revenue", axis=1)
y = data["Revenue"]

# Veri setini dengeleme (oversampling)
os = RandomOverSampler(sampling_strategy=0.85)
X_res, y_res = os.fit_resample(X, y)

# EÄŸitim ve test setlerine ayÄ±rma
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.33, random_state=42)

# SVM modelini tanÄ±mlama ve eÄŸitme
svm_model = SVC()
svm_model.fit(X_train, y_train)

# EÄŸitim doÄŸruluÄŸunu hesaplama
train_accuracy = svm_model.score(X_train, y_train)

# Test doÄŸruluÄŸunu hesaplama
test_accuracy = svm_model.score(X_test, y_test)

# SonuÃ§larÄ± yazdÄ±rma
print(f'EÄŸitim DoÄŸruluÄŸu: {train_accuracy}')
print(f'Test DoÄŸruluÄŸu: {test_accuracy}')

# Grafik Ã§izdirme
plt.figure(figsize=(8, 6))
sns.barplot(x=['EÄŸitim', 'Test'], y=[train_accuracy, test_accuracy])
plt.title('SVM Modeli EÄŸitim ve Test DoÄŸruluklarÄ±')
plt.ylabel('DoÄŸruluk')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler

# Veri setini yÃ¼kleme
file_path = r'C:\Users\sinem\Desktop\online_shoppers_intention.csv'
data = pd.read_csv(file_path)

# Kategorik verileri one-hot encoding ile sayÄ±sal verilere dÃ¶nÃ¼ÅŸtÃ¼rme
data = pd.get_dummies(data, columns=["Month", "VisitorType"], drop_first=True)

# Hedef deÄŸiÅŸken ve baÄŸÄ±msÄ±z deÄŸiÅŸkenleri ayÄ±rma
X = data.drop("Revenue", axis=1)
y = data["Revenue"]

# Veri setini dengeleme (oversampling)
os = RandomOverSampler(sampling_strategy=0.85)
X_res, y_res = os.fit_resample(X, y)

# EÄŸitim ve test setlerine ayÄ±rma
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.33, random_state=42)

# Modelleri tanÄ±mlama
models = {
    'SVM': SVC(),
    'Naive Bayes': GaussianNB(),
    'Lineer Regresyon': LogisticRegression(max_iter=10000),
    'Karar AÄŸacÄ±': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

# Modelleri eÄŸitme ve doÄŸruluk hesaplama
results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    train_accuracy = accuracy_score(y_train, y_pred_train)
    test_accuracy = accuracy_score(y_test, y_pred_test)
    results[model_name] = (train_accuracy, test_accuracy)
    print(f'{model_name} - EÄŸitim DoÄŸruluÄŸu: {train_accuracy}, Test DoÄŸruluÄŸu: {test_accuracy}')

# En iyi doÄŸruluk oranÄ±nÄ± bulma
best_model = max(results, key=lambda k: results[k][1])
best_accuracy = results[best_model][1]
print(f'En iyi model: {best_model}, Test DoÄŸruluÄŸu: {best_accuracy}')

# Grafik Ã§izdirme
plt.figure(figsize=(12, 8))
model_names = list(results.keys())
train_accuracies = [results[model][0] for model in model_names]
test_accuracies = [results[model][1] for model in model_names]

x = np.arange(len(model_names))

plt.bar(x - 0.2, train_accuracies, 0.4, label='EÄŸitim DoÄŸruluÄŸu')
plt.bar(x + 0.2, test_accuracies, 0.4, label='Test DoÄŸruluÄŸu')

plt.xlabel('Model')
plt.ylabel('DoÄŸruluk OranÄ±')
plt.title('Modellere GÃ¶re EÄŸitim ve Test DoÄŸruluk OranlarÄ±')
plt.xticks(x, model_names)
plt.legend()

plt.show()
```
