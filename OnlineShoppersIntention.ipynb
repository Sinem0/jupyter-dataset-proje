```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler

# Veri setini yükleme
file_path = r'C:\Users\sinem\Desktop\online_shoppers_intention.csv'
data = pd.read_csv(file_path)

# Kategorik verileri one-hot encoding ile sayısal verilere dönüştürme
data = pd.get_dummies(data, columns=["Month", "VisitorType"], drop_first=True)

# Hedef değişken ve bağımsız değişkenleri ayırma
X = data.drop("Revenue", axis=1)
y = data["Revenue"]

# Veri setini dengeleme (oversampling)
os = RandomOverSampler(sampling_strategy=0.85)
X_res, y_res = os.fit_resample(X, y)

# Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.33, random_state=42)

# SVM modelini tanımlama ve eğitme
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Eğitim doğruluğunu hesaplama
train_accuracy = svm_model.score(X_train, y_train)

# Test doğruluğunu hesaplama
test_accuracy = svm_model.score(X_test, y_test)

# Sonuçları yazdırma
print(f'Eğitim Doğruluğu: {train_accuracy}')
print(f'Test Doğruluğu: {test_accuracy}')

# Grafik çizdirme
plt.figure(figsize=(8, 6))
sns.barplot(x=['Eğitim', 'Test'], y=[train_accuracy, test_accuracy])
plt.title('SVM Modeli Eğitim ve Test Doğrulukları')
plt.ylabel('Doğruluk')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler

# Veri setini yükleme
file_path = r'C:\Users\sinem\Desktop\online_shoppers_intention.csv'
data = pd.read_csv(file_path)

# Kategorik verileri one-hot encoding ile sayısal verilere dönüştürme
data = pd.get_dummies(data, columns=["Month", "VisitorType"], drop_first=True)

# Hedef değişken ve bağımsız değişkenleri ayırma
X = data.drop("Revenue", axis=1)
y = data["Revenue"]

# Veri setini dengeleme (oversampling)
os = RandomOverSampler(sampling_strategy=0.85)
X_res, y_res = os.fit_resample(X, y)

# Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.33, random_state=42)

# Modelleri tanımlama
models = {
    'SVM': SVC(),
    'Naive Bayes': GaussianNB(),
    'Lineer Regresyon': LogisticRegression(max_iter=10000),
    'Karar Ağacı': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

# Modelleri eğitme ve doğruluk hesaplama
results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    train_accuracy = accuracy_score(y_train, y_pred_train)
    test_accuracy = accuracy_score(y_test, y_pred_test)
    results[model_name] = (train_accuracy, test_accuracy)
    print(f'{model_name} - Eğitim Doğruluğu: {train_accuracy}, Test Doğruluğu: {test_accuracy}')

# En iyi doğruluk oranını bulma
best_model = max(results, key=lambda k: results[k][1])
best_accuracy = results[best_model][1]
print(f'En iyi model: {best_model}, Test Doğruluğu: {best_accuracy}')

# Grafik çizdirme
plt.figure(figsize=(12, 8))
model_names = list(results.keys())
train_accuracies = [results[model][0] for model in model_names]
test_accuracies = [results[model][1] for model in model_names]

x = np.arange(len(model_names))

plt.bar(x - 0.2, train_accuracies, 0.4, label='Eğitim Doğruluğu')
plt.bar(x + 0.2, test_accuracies, 0.4, label='Test Doğruluğu')

plt.xlabel('Model')
plt.ylabel('Doğruluk Oranı')
plt.title('Modellere Göre Eğitim ve Test Doğruluk Oranları')
plt.xticks(x, model_names)
plt.legend()

plt.show()
```
